**Примеры индексации**

| Данные                            | Индексация                     |
|-----------------------------------|--------------------------------|
| Одна ячейка                       | .loc[7, 'genre']               |
| Один столбец                      | .loc[:, 'genre']               |
| Несколько столбцов                | .loc[:, ['genre', 'artist']]   |
| Несколько столбцов подряд (срез)  | .loc[:, 'total play': 'genre'] |
| Одна строка                       | .loc[1]                        |
| Все строки, начиная с заданной    | .loc[1:]                       |
| Все строки до заданной            | .loc[:3]                       |
| Несколько строк подряд (срез)     | .loc[2:5]                      |

**В логической индексации можно использовать логические выражения с любыми логическими операциями: ==,  !=, >, <, >=, <=**, а также функции-предикаты, которые возвращают логическое значение.

```python
mini_map[(mini_map['area_kv'] > 500) & (mini_map['population'] > 50000000)]
```

```python
mini_map[(mini_map['population'] >= 60000000) & (mini_map['area_kv'] <= 600)]
```

```python
mini_map[mini_map['country'] == 'Норвегия']
```

```python
# Функции-предикторы
# Поиск страны, название которой начинается с буквы 'Н'

mini_map[mini_map['country'].str.startswith('Н')]
```

```python
# Поиск страны, в названии которой есть буква 'а'

mini_map[mini_map['country'].str.contains('а')]
```

```python
# Поиск страны, у которой население не является числом
# (исключительно числовой тип данных)

mini_map[~mini_map['population'].apply(lambda x: isinstance(x, int))]
```
